```python
def gradientDescent(gradient, start=0, rate =1, iterations =100):
    x_old = start
    for i in range(iterations):
        x_new = x_old -rate * gradient(x_old)
        if(x_new ==  x_old):
            break 
        x_old = x_new
    return x_new
```

```python
def g(x):
    return 4*x**3 + 3*x**2 - 4*x
```

```python
gradientDescent(g,-1,0.1,500)
```
